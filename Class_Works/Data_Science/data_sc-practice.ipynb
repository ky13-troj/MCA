```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = np.linspace(0, 2**np.pi, 400)
y = np.sin(x**2)
fig = plt.figure() # create a figure object
ax = fig.add_subplot(1, 1, 1) # create an axes object in the figure
ax.plot(x, y) # plot the data
plt.show()
```

```python
# Creating the Discrete Plot
x_dis = x[::20]
y_dis = y[::20]

'''
[::20] means every 20th element of the array will be taken.
'''

plt.stem(x_dis, y_dis, markerfmt='ro', linefmt='b-', basefmt='g-')

''' 
stem is used to create the discrete plot with markers 
markerfmt is used to specify the marker format
linefmt is used to specify the line format
basefmt is used to specify the base format
'''
#Customizing the plot
plt.title("Discrete Plot")
plt.xlabel("x")
plt.ylabel("y")

#Display the Plot
plt.show()
```

```python
# Printing the x discrete values as DataFrame:
x_data = pd.DataFrame(x_dis[:10], columns=['x']) # create a DataFrame with 10 [first 10 data points of x_dis array]rows and 1 column
```

```python
# Sub - Plot

x= np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1,color='black')

plt.subplot(2, 2, 2)
plt.plot(x, y2,color='blue')

plt.subplot(2, 2, 3)
plt.plot(x, y3,color='red')       

plt.subplot(2, 2, 4)
plt.plot(x, y3,color='green')   
```

```python
frequency = 10
sampling_rate = 100
duration = 1.0

t = np.arange(0, duration, 1/sampling_rate)
sin_wave = np.sin(2*np.pi*frequency*t)

plt.figure(figsize=(40, 30))
plt.subplot(2, 2, 1)
plt.stem(t, sin_wave, markerfmt='ro', linefmt=':', basefmt='b')

plt.subplot(2, 2, 2)
sampling_rate = 1000
t = np.arange(0, duration, 1/sampling_rate)
sin_wave = np.sin(2 * np.pi * frequency * t)
plt.stem(t, sin_wave, markerfmt='ro', linefmt='-')

plt.xlabel("Time")
plt.ylabel("amplitude")
plt.grid(True) # adds grid
# here plot function is not used because it connects the discrete data points together
plt.show()
```
