```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

# Plotting the data
plt.plot(x_values, y_values, label='Sample Data', marker='o')

# Adding labels to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding a title to the plot
plt.title('Simple Plot Example')

# Adding a legend to the plot
plt.legend()

# Displaying the plot
plt.show()
```

```python
#subplot

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Sample data

x = np.linspace(0, 2**np.pi, 400)
y = np.sin(x**2)
fig = plt.figure()
ax = fig.subplots()
ax.plot(x, y)
fig.show()
```

```python
x = np.linspace(0, 2**np.pi, 400)
y = np.sin(x**2)

x_dis = x[::20]
y_dis = y[::20]
plt.stem(x_dis, y_dis, markerfmt='ro', linefmt='b-',basefmt='-')

# Customize plot
plt.title('Discrete Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
import pandas as pd
# print(f"x Values :\n{x}")
x_data = pd.DataFrame(x_dis)
print(x_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1,color='black')

plt.subplot(2, 2, 2)
plt.plot(x, y2,color='blue')

plt.subplot(2, 2, 3)
plt.plot(x, y3,color='red')       

plt.subplot(2, 2, 4)
plt.plot(x, y3,color='green')   
```

```python
frequency = 10
sampling_rate = 100
duration = 1.0

t = np.arange(0, duration, 1/sampling_rate)

# y = np.sin(2 * np.pi * frequency * t)

sin_wave = np.sin(2 * np.pi * frequency * t)

plt.figure(figsize=(8, 8))

# plt.plot(t, sin_wave)
plt.stem(t, sin_wave, markerfmt='ro', linefmt=':')
plt.xlabel("Time")
plt.ylabel("amplitude")
plt.grid(True)
plt.show()
```
